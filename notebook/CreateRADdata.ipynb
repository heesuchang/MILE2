##  This notebook allows the user to refine the data.csv created in Metadata2Data by the Evaluator service. The resulting csv can be copied into a Recommedations Analysis Dashboard to analyze the collection for recommendation completeness for about 30 different Earth and Space Science communities recommendations. 

Import needed python packages

```python
import pandas as pd
pd.options.display.max_colwidth=200
import os
from os import walk
from ipywidgets import *
import ipywidgets as widgets
```

Create a list of data.csv that are in the MILE2 repository

```python
DataFiles=[]
for dirpath, dirnames, filenames in os.walk("../data"):
    for filename in [f for f in filenames if f.endswith(".csv")]:
        DataFiles.extend([os.path.join(dirpath, filename)])
DataFiles  
```

Create a function that allows us to import the csv contents into a dataframe.

```python
def DataChoices(DataFile):
    global CollectionConceptsDF
    CollectionConceptsDF= pd.read_csv(os.path.join(DataFile))
    return CollectionConceptsDF
```

Create dropdown to select csv

```python
interactive(DataChoices,DataFile=DataFiles)
```

Drop the columns we don't need in the RAD

```python
#CollectionConceptsDF.drop(['XPath','Content','DialectDefinition'], axis=1, inplace=True)
CollectionConceptsDF
```

Remove duplicate rows to increase processing speed in Excel.

```python
ConceptsDF=CollectionConceptsDF.drop_duplicates()
```

Create a new csv with just the information we want in it

```python
ConceptsDF.to_csv('../data/dataForRAD.csv', mode = 'w', index=False)
```
