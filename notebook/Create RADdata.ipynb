```python
def fileChoices(Record):
    global OrganizationChoice
    global Collections
    global HDFconceptsDF
    dataPath=os.path.join('../HDF',Record)
    HDFconceptsDF=pd.read_hdf(dataPath, key=Record[:-3])
    return HDFconceptsDF
```

```python
interactive(fileChoices,Record=h5)
```
